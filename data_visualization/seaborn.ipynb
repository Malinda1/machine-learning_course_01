```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.displot(tips['total_bill'], kde = True, bins = 30)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'reg')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde', cmap = 'viridis') # We can use cmap for difference type of color map 'plasma', 'inferno', 'magma', 'cividis'
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = 'sex')
```

```python
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

# Visualize categorical features

```python
 sns.barplot(x = 'sex', y = 'total_bill', data = tips, hue = 'sex', palette = 'Set2')
```

```python
import numpy as np
sns.barplot(x = 'sex', y = 'total_bill', data = tips, hue = 'sex', palette = 'Set2', estimator = np.std)
```

```python
sns.countplot(x = 'sex', data = tips, hue = 'sex', palette = 'Set1')
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker')
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True)
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = True)
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = True, hue = 'sex')
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = True, hue = 'sex', dodge = True)
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'day')
sns.swarmplot(x = 'day', y = 'total_bill', data = tips, color = 'black')
```

# matrix visualization

```python
flight = sns.load_dataset('flights')
flight.head()
```

```python
numeric_tips = tips.select_dtypes(include = ['float64', 'int64'])

tc = numeric_tips.corr()
sns.heatmap(tc, cmap = 'coolwarm', annot = True )
```

```python
tc
```

```python
fp = flight.pivot_table(index = 'month', columns = 'year', values = 'passengers', observed = False)
plt.figure(figsize = (8, 7))
sns.heatmap(fp, cmap = 'coolwarm', linecolor = 'black', linewidths = 0.5)
```

```python
sns.clustermap(fp, linecolor = 'white', linewidths = 0.5)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
Grid = sns.PairGrid(iris)
Grid.map_diag(sns.histplot)
Grid.map_upper(sns.scatterplot)
Grid.map_lower(sns.kdeplot)
Grid.add_legend()
```
