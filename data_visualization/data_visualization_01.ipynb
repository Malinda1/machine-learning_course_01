```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Basic plot (functional method)

plt.plot(x, y)
```

```python
# Adding title, xlabel and ylabel

plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# Create a multiple plot same canvas

plt.subplot(1, 2, 1)
plt.plot(x, y, 'g')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'r')
```

```python
# Object oriented method

fig = plt.figure(figsize = (10, 7))

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
```

```python
# set the label and title

fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
# Create a mupltiple axes 

fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.4, 0.2, 0.4, 0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Define the location of plot
axes2 = fig.add_axes([0.15, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title('Large Plot')
axes2.plot(y, x)
axes2.set_title('Small Plot')
```

```python
# Create a sub plot using object oriented

fig, axes = plt.subplots(nrows = 1, ncols = 2)

for current_ax in axes:
    current_ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x, y)
axes[0].set_title('First plot')

axes[1].plot(y, x)
axes[1].set_title('Second plot')

plt.tight_layout()
```

```python
# Figure Size and DPI

fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (6, 6))

axes[0].plot(x, y)
axes[1].plot(y, x)
```

```python
# Save the plot 

fig.save
```
