```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Define column names
columns_names = ['user_id', 'item_id', 'rating', 'timestamp']

# Read the data from the file 'u.data'
data_01 = pd.read_csv(
    'u.data',  # File path
    sep='\t',  # Separator (tab-separated file)
    names=columns_names  # Column names
)

data_01.head()
```

```python
movies_titles = pd.read_csv('Movie_Id_Titles')
```

```python
data_02 = pd.merge(data_01, movies_titles, on = 'item_id')

data_02.head()
```

```python
data_02.groupby('title')['rating'].mean().sort_values(ascending = False).head()
```

```python
data_02.groupby('title')['rating'].count().sort_values(ascending = False).head()
```

```python
ratings = pd.DataFrame(data_02.groupby('title')['rating'].mean())
ratings['num_of_ratings'] = pd.DataFrame(data_02.groupby('title')['rating'].count())

ratings.head()
```

```python
ratings['num_of_ratings'].hist(bins = 70)
```

```python
ratings['rating'].hist(bins = 70)
```

```python
sns.jointplot(
    x = 'rating',
    y = 'num_of_ratings',
    data = ratings,
    alpha = 0.7
)
```

```python
moviemat = data_02.pivot_table(index = 'user_id', columns = 'title', values = 'rating')

moviemat.head()
```

```python
ratings.sort_values(
    'num_of_ratings',
    ascending = False
).head(10)
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
liarlair_user_rating = moviemat['Liar Liar (1997)']
```

```python
starwars_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
```

```python
similar_to_liarlair = moviemat.corrwith(liarlair_user_rating)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns = ['Correlation'])
corr_starwars.dropna(inplace = True)

corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation', ascending = False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['num_of_ratings'])

corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num_of_ratings'] > 100].sort_values('Correlation', ascending = False).head()
```

```python
corr_lairlair = pd.DataFrame(similar_to_liarlair, columns = ['Correlation'])
corr_lairlair.dropna(inplace = True)

corr_lairlair.head()
```

```python
corr_lairlair = corr_lairlair.join(ratings['num_of_ratings'])

corr_lairlair.head()
```

```python
corr_lairlair[corr_lairlair['num_of_ratings'] > 100].sort_values('Correlation', ascending = False).head()
```
